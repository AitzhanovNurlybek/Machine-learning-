```python
#4.1
import numpy as np
import matplotlib as plt 

np.random.seed(42)
Y = np.concatenate([np.random.normal(2.0, 1.0, size=150), np.random.normal(6.0, 1.5, size=150)])
len(Y), Y[:10]
```

```python
#4.2
c_values = np.linspace(-1, 10, 500)
mse_values = [np.mean((Y - c)**2) for c in c_values]
```

```python
#4.3
opt_idx = np.argmin(mse_values)
opt_c = c_values[opt_idx]
mean_Y = Y.mean()
opt_c, mean_Y
```

```python
#4.4
import matplotlib.pyplot as plt 

plt.figure(figsize=(7,4))
plt.plot(c_values, mse_values, label="MSE(c)")
plt.axvline(opt_c, color='r', linestyle='--', label=f'Optimal c={opt_c:.2f}')
plt.scatter([opt_c], [mse_values[opt_idx]], color='red')
plt.xlabel("c")
plt.ylabel("MSE")
plt.title("MSE vs c")
plt.legend()
plt.grid(True)
plt.show()
```

#4.5
The curve is U-shaped parabola minimum is at the mean of Y the further c moves away from mean(Y) the bigger the MSE
